```python
import pandas as pd
from snorkel.labeling import LabelingFunction
from snorkel.labeling import PandasLFApplier
from snorkel.labeling import LFAnalysis
from nltk.tokenize import sent_tokenize
#, word_tokenize
```

```python
import pandas as pd
```

```python
#Supply a list of reviews/customer feedback
df=pd.read_csv("ChromeRevs5000.csv",header=None)
df.columns=["no","reviews"]
len(df)
```

```python
dfs = []
df=df.drop("no",axis=1)
df.columns = map(str.lower, df.columns)
df = df.sample(frac=1, random_state=123).reset_index(drop=True)
dfs.append(df)
df_train = pd.concat(dfs[:4])
len(df_train)
```

```python
sents=[]
for rev in df_train['reviews']:
    sentences=sent_tokenize(rev)
    sents=sents+sentences
#     for s in sentences:
#         sents=sents+[s]
    
    
len(sents)
```

```python
df_train=pd.DataFrame(sents,columns=['reviews'])
print(df_train.shape)
df_train['reviews']=df_train['reviews'].str.lower()
df_train.head(5)
```

```python
from snorkel.labeling import LabelingFunction

def keyword_lookup(x, keywords, label):
    if any(word in x.reviews.lower() for word in keywords):
        return label
    return ABSTAIN


ABSTAIN = -1
OPTICAL_DRIVE=1
DISPLAY = 2
MEMORY = 3
BATTERY=4
FAN = 5
MOTHERBOARD = 6
HARDDRIVE = 7
SHIPPING = 8
CPU = 9
CHARGER = 10
SUPPORT = 11
KEYBOARD = 12
OS=13
LAPTOP=14
PERFORMANCE=15
CONNECTIVITY=16
MULTIMEDIA=17

#     keyword_optical_drive,
#     keyword_display,
#     keyword_memory,
#     keyword_battery,
#     keyword_fan,
#     keyword_motherboard,
#     keyword_harddrive,
#     keyword_shipping,
#     keyword_cpu,
#     keyword_charger,
#     keyword_support,
#     keyword_keyboard,
#     keyword_operating_system,
#     keyword_laptop,

###Optical drives feature
def make_keyword_optical_lf(keywords, label=OPTICAL_DRIVE):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )


#"""reviews taking about optical drives have following keywords: cd,dvd,cd drive dvd drive"""
keyword_optical_drive = make_keyword_optical_lf(keywords=["optical drive","optical drives",
                                                          "dvd","dvd drive","dvd/cd",
                                                          "cd_rom_drive","cd rom",
                                                          "cd rom drive","burner",
                                                          "videodisk","videodisc","superdrive",
                                                          "compact disk", "compact disc"])



##Display feature
def make_keyword_display_lf(keywords, label=DISPLAY):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_display =  make_keyword_display_lf(keywords=["display","IPS","HD","picture quality","picture","monitor", 
                                                     "screen","touch screen", "touchscreen", 
                                                     "pixel","lcd", "resolution","video","11.6" ,
                                                     "12.5","13.3","14","15.6", "17.3", "18.4","11.6 inches" ,
                                                     "12.5 inches","13.3 inches","14 inches",
                                                     "15.6 inches", "17.3 inches", "18.4 inches",
                                                     "11.6\"" ,"12.5\"","13.3\"","14\"","15.6\"", "17.3\"", "18.4\""])


##Memory feature
def make_keyword_memory_lf(keywords, label=MEMORY):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_memory =  make_keyword_memory_lf(keywords=["memory","memory chip","ram","random access memory", "memory board", "memory card","memory stick"]) 


##battery feature
def make_keyword_battery_lf(keywords, label=BATTERY):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_battery =  make_keyword_battery_lf(keywords=["battery","recharging","battries",
                                                     "recharge","power unit","power_unit",
                                                     "electric cell","electric_cell","cell","cells",
                                                     "electric_battery","power cell"]) 






#FANS feature 
def make_keyword_fan_lf(keywords, label=FAN):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_fan =  make_keyword_fan_lf(keywords=["fan","cooling system","cpu fan","fans","heating",
                                             "cooling","heat up","cool down","cool off","temperature","temperature reduction"]) 





#Mother board feature 
def make_keyword_motherboard_lf(keywords, label=MOTHERBOARD):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_motherboard =  make_keyword_motherboard_lf(keywords=["motherboard","mother board","mother-board","mother_board","cpu-board","cpu board","cpu_board"]) 





#hard disk feature 
def make_keyword_harddrive_lf(keywords, label=HARDDRIVE):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_harddrive =  make_keyword_harddrive_lf(keywords=["harddrive","hard drive",
                                                         "hard drives","disc","disk",
                                                         "hard disc","hard disc","solid state",
                                                         "solid-state","hdd","ssd","magnetic disc",
                                                         "magnetic disc"]) 



#sHIPPING feature 
def make_keyword_shipping_lf(keywords, label=SHIPPING):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_shipping =  make_keyword_shipping_lf(keywords=["shipping","delivery"])




#cpu feature 
def make_keyword_cpu_lf(keywords, label=CPU):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_cpu =  make_keyword_cpu_lf(keywords=["cpu","processing unit","processor", "central processing unit","i3","i5","i7"])







#charger feature 
def make_keyword_charger_lf(keywords, label=CHARGER):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_charger =  make_keyword_charger_lf(keywords=["charger","battery charger", "battery-charger"])



#support feature 
def make_keyword_support_lf(keywords, label=SUPPORT):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_support =  make_keyword_support_lf(keywords=["support","tech support","tech-support","technical support","call centre","call center",
 "contact center","contact centre","answering service", "answering services"
,"helpline","hotline","troubleshooting","troubleshoo","help desk","help line","after-sale service","service"])







#keyboard feature 
def make_keyword_keyboard_lf(keywords, label=KEYBOARD):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_keyboard =  make_keyword_keyboard_lf(keywords=["keyboard","typing","keys","key boards","key board"])



##keyword_operating_system fearure


#keyboard feature 
def make_keyword_operating_system_lf(keywords, label=OS):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_operating_system =  make_keyword_operating_system_lf(keywords=["os","operating system","operating_system","window 8","windows",
                                                                       "linux","mac os","ubuntu","fedora","chrome os"])



#laptop feature 
def make_keyword_laptop_lf(keywords, label=LAPTOP):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_laptop =  make_keyword_laptop_lf(keywords=["laptop","chromebook","computer","pc","macbook",
                                                   "chromebook ","chromebooks","machine","product"])




#performance feature 
def make_keyword_performance_lf(keywords, label=PERFORMANCE):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_performance =  make_keyword_performance_lf(keywords=["performance","speed","fast","slow"])



#performance feature 
def make_keyword_connectivity_lf(keywords, label=CONNECTIVITY):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_connectivity =  make_keyword_connectivity_lf(keywords=["connectivity","rj-45","rj45","ethernet", 
                                                               "ethernet adapter",
                                                               "ethernet port","wired network connection", 
                                                                 "network connection", "bluetooth","wireless",
                                                               "wifi","wi-fi","wan","lan","usb","port","ports",
                                                               "connector","connecter"])


#multimedia e feature 
def make_keyword_multimedia_lf(keywords, label=MULTIMEDIA):
    return LabelingFunction(
        name=f"keyword_{keywords[0]}",
        f=keyword_lookup,
        resources=dict(keywords=keywords, label=label),
    )

keyword_multimedia =  make_keyword_multimedia_lf(keywords=["multimedia","sound","audio","aux","audio jack",
                                                           "loud","speaker","headphone","earphone","headset","headphones",
                                                           "earphones","head phone","ear phone","head set",
                                                           "head phones","ear phones","mp3","mp4","mpeg","midi"])



```

```python
lfs = [
    keyword_optical_drive,
    keyword_display,
    keyword_memory,
    keyword_battery,
    keyword_fan,
    keyword_motherboard,
    keyword_harddrive,
    keyword_shipping,
    keyword_cpu,
    keyword_charger,
    keyword_support,
    keyword_keyboard,
    keyword_operating_system,
    keyword_laptop,
    keyword_performance,
    keyword_connectivity,
    keyword_multimedia,]

applier = PandasLFApplier(lfs=lfs)


L_train = applier.apply(df=df_train)
```
